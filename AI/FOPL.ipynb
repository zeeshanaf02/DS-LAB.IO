```python
from sympy import symbols, Or, Not

def resolution(clause1, clause2):
    resolved = False
    new_clause = None
    
    # Convert clause1 and clause2 to their arguments if they are Or objects
    if isinstance(clause1, Or):
        clause1 = clause1.args
    else:
        clause1 = [clause1]
        
    if isinstance(clause2, Or):
        clause2 = clause2.args
    else:
        clause2 = [clause2]
    
    for literal2 in clause2:
        if isinstance(literal2, Not):
            negated_literal = literal2.args[0]  # Get the negated literal
            if negated_literal in clause1:
                resolved = True
                new_clause = Or(*[l for l in clause1 if l != negated_literal] + [l for l in clause2 if l != literal2])
                break
    return resolved, new_clause

def resolution_algorithm(clauses):
    new_clauses = set(clauses)
    while True:
        new_clauses_size = len(new_clauses)
        for clause1 in list(new_clauses):
            for clause2 in list(new_clauses):
                if clause1 != clause2:
                    resolved, new_clause = resolution(clause1, clause2)
                    if resolved and new_clause is not None and new_clause not in new_clauses:
                        new_clauses.add(new_clause)
            if len(new_clauses) == new_clauses_size:
                break
            new_clauses_size = len(new_clauses)
    return new_clauses

if __name__ == "__main__":
    A, B, C, D = symbols('A B C D')
    clauses = [
        Or(A, B, C),
        Or(Not(A), D),
        Or(Not(B), Not(C)),
        Or(Not(D))
    ]
    new_clauses = resolution_algorithm(clauses)
    print("Resulting Clauses after Resolution:")
    for clause in new_clauses:
        print(clause)
```
